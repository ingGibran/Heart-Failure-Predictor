# Predicción falla cardiaca

## 1. Cargar Datos

```python
import pandas as pd

df = pd.read_csv('heart_failure.csv')
df
```

## 2. Limpiar

```python
# dividir datos
from sklearn.model_selection import train_test_split

X = df.drop('DEATH_EVENT', axis=1)
y = df['DEATH_EVENT']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=77)
```

```python
# Escalar
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

vars = ['age', 'creatinine_phosphokinase', 'ejection_fraction', 'platelets', 'serum_creatinine', 'serum_sodium', 'time']
X_train[vars] = scaler.fit_transform(X_train[vars])
X_test[vars] = scaler.fit_transform(X_test[vars])
X_train
```

## 3. Modelar

```python
from sklearn import svm

model = svm.SVC()
model.fit(X_train, y_train)
```

## 4. Evaluar

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = model.predict(X_test)

print( f"Accuracy: {accuracy_score(y_test, y_pred)}" )
print( classification_report(y_test, y_pred) )
print( confusion_matrix(y_test, y_pred) )
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Asumiendo que ya hiciste model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# 1. Accuracy (¿Qué porcentaje atinó?)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")

# 2. Reporte completo (Importante para datos médicos)
# Fíjate en el "Recall" del caso 1 (Muerte), es vital en medicina.
print(classification_report(y_test, y_pred))

# 3. Matriz (Para ver falsos positivos/negativos)
print(confusion_matrix(y_test, y_pred))
```
